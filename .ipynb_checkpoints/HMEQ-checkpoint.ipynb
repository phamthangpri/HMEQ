# Context

The consumer credit department of a bank wants to automate the decisionmaking process for approval of home equity lines of credit. To do this, they will follow the recommendations of the Equal Credit Opportunity Act to create an empirically derived and statistically sound credit scoring model. The model will be based on data collected from recent applicants granted credit through the current process of loan underwriting. The model will be built from predictive modeling tools, but the created model must be sufficiently interpretable to provide a reason for any adverse actions (rejections).


Link Kaggle : https://www.kaggle.com/datasets/ajay1735/hmeq-data

The Home Equity dataset (HMEQ) contains baseline and loan performance information for 5,960 recent home equity loans. The target (BAD) is a binary variable indicating whether an applicant eventually defaulted or was seriously delinquent. This adverse outcome occurred in 1,189 cases (20%). For each applicant, 12 input variables were recorded.

__https://www.kaggle.com/code/akhil14shukla/loan-defaulter-prediction__

https://www.kaggle.com/code/ajay1735/my-credit-scoring-model/notebook

# Approach

+ In this analysis,  we will test different model (by gridsearchcv) with the original data (data leakage).
+ Then we will try 2 methods different to process data leakage : undersampling and oversampling. 2 models the most performant from the first step will be used for this section.

In the end, we will also use neural network to know if the result is better

```python
def remove_outliers_iqr(col):
    Q1 = df_data[col].quantile(0.25)
    Q3 = df_data[col].quantile(0.75)
    IQR = Q3 - Q1
    S = 1.5*IQR
    LB = Q1 - S
    UB = Q3 + S
    df_data.loc[df_data[col] > UB,col] = UB
    df_data.loc[df_data[col] < LB,col] = LB
    return df_data
```

# Preproc

```python
df_data.info()
```

```python
y = df_data["BAD"]
X = df_data.drop(columns="BAD")
```

```python
preproc_numerical_baseline = make_pipeline(
    SimpleImputer(strategy = "mean"),
    MinMaxScaler())

preproc_categorical_baseline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(drop = "if_binary", handle_unknown="ignore"))

preproc_transform = make_column_transformer(
    (preproc_numerical_baseline, make_column_selector(dtype_include=["int64", "float64"])),
    (preproc_categorical_baseline, make_column_selector(dtype_include=["object"])),
    remainder="drop")
preproc_transform
```

```python
X_transform = preproc_transform.fit_transform(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)
```

```python
list_results = []
def simulate_model(model,model_name, list_results):
    ######### Add model #############
    pipeline_ensemble = make_pipeline(preproc_transform, model)

    y_pred = cross_val_predict(pipeline_ensemble, X_train, y_train, cv=5)

    accuracy = round(accuracy_score(y_train, y_pred), 2)
    recall = round(recall_score(y_train,y_pred),2)
    precision = round(precision_score(y_train, y_pred),2)
    F1_Score = round(f1_score(y_train,y_pred),2)

    print(f'accuracy : {accuracy}')
    print(f'recall : {recall}')
    print(f'precision : {precision}')
    print(f'F1_Score : {F1_Score}')
    list_results.append([model_name,accuracy,recall,precision,F1_Score])
    return list_results
```

```python
def print_results(pipe_ensemble,X_test,y_test):  
    y_pred_proba = pipeline_ensemble.predict(X_test)
    threshold = 0.50 # 50% 

    y_pred = np.where(y_pred_proba > threshold, 1, 0 )

    accuracy = round(accuracy_score(y_test, y_pred), 4)
    recall = round(recall_score(y_test,y_pred),4)
    precision = round(precision_score(y_test, y_pred),4)
    F1_Score = round(f1_score(y_test,y_pred),4)

    print(f'accuracy : {accuracy}')
    print(f'recall : {recall}')
    print(f'precision : {precision}')
    print(f'F1_Score : {F1_Score}')

    cm = confusion_matrix(y_test,y_pred)

    plt.figure(figsize=(5,5))
    sns.heatmap(cm, annot=True, fmt="d")
    plt.title('Confusion matrix')
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
    plt.show()
```

## Oversampling 

```python
def plot_loss_precision_recall_curve(history):
    
    fig, ax =plt.subplots(1,3,figsize=(20,5))
    
    # --- LOSS
    
    ax[0].plot(history.history['loss'])
    ax[0].plot(history.history['val_loss'])
    ax[0].set_title('Model loss')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')
    ax[0].legend(['Train', 'Val'], loc='upper right')
    ax[0].grid(axis="x",linewidth=0.5)
    ax[0].grid(axis="y",linewidth=0.5)
    
    
    # --- RECALL

    ax[1].plot(history.history['recall'])
    ax[1].plot(history.history['val_recall'])
    ax[1].set_title('Model recall', fontsize = 18)
    ax[1].set_xlabel('Epoch', fontsize = 14)
    ax[1].set_ylabel('Recall', fontsize = 14)
    ax[1].legend(['Train', 'Val'], loc='lower right') 
    ax[1].grid(axis="x",linewidth=0.5)
    ax[1].grid(axis="y",linewidth=0.5)

    
    # --- PRECISION
    
    
    ax[2].plot(history.history['precision'])
    ax[2].plot(history.history['val_precision'])
    ax[2].set_title('Model precision', fontsize = 18)
    ax[2].set_xlabel('Epoch', fontsize = 14)
    ax[2].set_ylabel('Precision', fontsize = 14)
    ax[2].legend(['Train', 'Val'], loc='lower right')  
    ax[2].grid(axis="x",linewidth=0.5)
    ax[2].grid(axis="y",linewidth=0.5)    
    
    plt.show()
```

```python
def print_results2(model,X_test,y_test):
    y_pred_proba = model.predict(X_test)
    threshold = 0.50 # 50% 

    y_pred = np.where(y_pred_proba > threshold, 1, 0 )

    accuracy = round(accuracy_score(y_test, y_pred), 4)
    recall = round(recall_score(y_test,y_pred),4)
    precision = round(precision_score(y_test, y_pred),4)
    F1_Score = round(f1_score(y_test,y_pred),4)

    print(f'accuracy : {accuracy}')
    print(f'recall : {recall}')
    print(f'precision : {precision}')
    print(f'F1_Score : {F1_Score}')

    cm = confusion_matrix(y_test,y_pred)

    plt.figure(figsize=(5,5))
    sns.heatmap(cm, annot=True, fmt="d")
    plt.title('Confusion matrix')
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
def plot_roc_curve(fpr, tpr):
    plt.plot(fpr, tpr, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()

y_true = np.array([0,0, 1, 1,1])
y_scores = np.array([0.0,0.09, .05, .75,1])

fpr, tpr, thresholds = roc_curve(y_true, y_scores)
print(tpr)
print(fpr)
print(thresholds)
print(roc_auc_score(y_true, y_scores))
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]
print("Threshold value is:", optimal_threshold)
plot_roc_curve(fpr, tpr)
```
